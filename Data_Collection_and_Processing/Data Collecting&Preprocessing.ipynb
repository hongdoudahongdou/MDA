# 1. Predict total deaths casued by heatwaves.

## Heatwave Data(EMDAT)

```python
df_heatwave = pd.read_csv("https://github.com/hongdoudahongdou/MDA/blob/main/data/heatwave.csv?raw=true")
df_heatwave.head()
```

This is a dataset of worldwide heatwaves from 1990 to 2023 collected from EMDAT. 

```python
#choose useful columns from emdat data
df_heatwave=df_heatwave[["Year","Country","ISO","Associated Dis","Associated Dis2","Appeal","Declaration","Start Year","Start Month","Start Day","End Year",
      "End Month","End Day","Total Deaths"]]
```

## GDP data 

```python
df_gdp=pd.read_csv('https://github.com/hongdoudahongdou/MDA/blob/main/data/gdp.csv?raw=true')
df_gdp.head()
```

This is a dataset of worldwide GDP data from 1960 to 2022 collected from the world bank. When merging the two datasets we will select the year from 1990 to 2022.

```python
#select the data from 1990 to 2021
df_gdp.drop(df_gdp.columns[2:df_gdp.columns.get_loc('1990')], axis=1,inplace=True)
df_gdp.head()
```

```python
#transform to melt table
df_gdp=pd.melt(df_gdp, id_vars=['Country Code'], value_vars=df_gdp.columns[2:35], 
            var_name='Year',value_name='GDP(million dollars)', col_level=None)
df_gdp.head()
```

## Merge Heatwave data and GDP data

```python
df_gdp.rename(columns={'Country Code': 'ISO'}, inplace=True)
df_gdp['Year'] = df_gdp['Year'].astype('int64')
merge_gdp = pd.merge(df_heatwave, df_gdp, on=['Year','ISO'],how='left')
merge_gdp.head()
```

```python
#drop the year 2023
merge_gdp = merge_gdp[merge_gdp['Year'] != 2023]
merge_gdp
```

## Merge with the population data

```python
df_population= pd.read_csv('https://github.com/hongdoudahongdou/MDA/blob/main/data/population.csv?raw=true')
df_population.head()
```

Above is a dataset of worldwide GDP data from 1960 to 2022 collected from the world bank. Again when merging thedatasets we will select the year from 1990 to 2022.

```python
#select the year from 1990 to 2022
df_population=df_population.rename(columns={'Country Code':'ISO'})
df_population.drop(df_population.columns[2:df_population.columns.get_loc('1990')], axis=1,inplace=True)
df_population.head()
```

```python
#transform to the melt table
df_population=pd.melt(df_population, id_vars=['ISO'], value_vars=df_population.columns[2:35], 
            var_name='Year',value_name='Population', col_level=None)
df_population.head()
```

```python
#Merge the data
df_population['Year']=df_population['Year'].astype(int)
merge_population = pd.merge(merge_gdp, df_population, on=['Year','ISO'],how='left')
merge_population
```

## Merge with the health expenditure data

```python
df_health=pd.read_csv('https://github.com/hongdoudahongdou/MDA/blob/main/data/health_expenditure.csv?raw=true')
df_health.head()
```

Above is a dataset of worldwide health expenditure data from 1960 to 2022 collected from the world bank. This dataset records the proportion of health expenditure to the GDP in various countries each year. Again, we select the years from 1990 to 2022.

```python
#select the year from 1990 to 2022
df_health=df_health.rename(columns={'Country Code':'ISO'})
df_health.drop(df_health.columns[2:df_health.columns.get_loc('1990')], axis=1,inplace=True)
df_health.head()
```

Note that most of the data of the year 2021 and 2022 are not available, so we will select the year from 1990 to 2020.

```python
#transform to the melt table and merge the data
df_health=pd.melt(df_health, id_vars=['ISO'], value_vars=df_health.columns[2:35], 
            var_name='Year',value_name='Health expenditure (% of GDP)', col_level=None)
df_health['Year']=df_health['Year'].astype(int)
merge_health = pd.merge(merge_population, df_health, on=['Year','ISO'],how='left')
merge_health
```

## Dealing with missing data

```python
merge_health.isna().sum()
```

### Replace the missing Associated Dis, Associated Dis 2, Appeal and Declaration with 0.

```python
merge_health=merge_health.fillna({"Associated Dis":0,"Associated Dis2":0,"Appeal":0,"Declaration":0})
merge_health.head()
```

```python
# Summary the type of the Associated Dis
dis_1=merge_health['Associated Dis'].unique()
dis_2=merge_health['Associated Dis2'].unique()
print(dis_1,dis_2)
```

```python
#Create new col for "Associated Drought" and "Associated Wildfire" 
assdrought = ((merge_health["Associated Dis"] == "Drought")|(merge_health["Associated Dis2"] == "Drought")).astype("int64")
asswildfire = ((merge_health["Associated Dis"] == "Wildfire")|(merge_health["Associated Dis2"] == "Wildfire")).astype("int64")
merge_health["Associated Drought"] = pd.Series(assdrought, name = 'Associated Drought')
merge_health["Associated Wildfire"] = pd.Series(asswildfire, name = 'Associated Wildfire')
merge_health = merge_health.drop(columns=["Associated Dis", "Associated Dis2"])
merge_health.head()
```

```python
# Create a new col for "Appeal or Declaration"
appl_dclr = ((merge_health["Appeal"] == "Yes")|(merge_health["Declaration"] == "Yes")).astype("int64")
merge_health["Appeal or Declaration"] = pd.Series(appl_dclr, name="Appeal or Declaration")
merge_health = merge_health.drop(columns=["Appeal", "Declaration"])
merge_health.head()
```

### Replace the missing Start Day and End Day with 1.

```python
merge_health=merge_health.fillna({"Start Day":1,"End Day":1})
merge_health.head()
```

### For the missing value of total deaths, GDP and healthexp , we drop these records.

```python
merge_health=merge_health.dropna()
merge_health
```

```python
#Create a new col 'Duration'
start = pd.DataFrame({'Year': merge_health['Start Year'],
                   'Month': merge_health['Start Month'],
                   'Day': merge_health['Start Day']})
start = pd.to_datetime(start)
end = pd.DataFrame({'Year': merge_health['End Year'],
                   'Month': merge_health['End Month'],
                   'Day': merge_health['End Day']})
end = pd.to_datetime(end)
duration = end-start
duration =duration.map(lambda x: x.days)+1
merge_health['duration']=duration
merge_health.head()
```

## Collect temperature data

```python
# transform iso_3 to iso_2 to use meteostat
merge_health['ISO_2']=merge_health['ISO'].apply(lambda x: pc.country_alpha3_to_country_alpha2(x))
merge_health.head()
```

```python
#transform float to int
merge_health['Start Day'] = merge_health['Start Day'].astype(int)
merge_health['End Day'] = merge_health['End Day'].astype(int)
```

```python
#create a new col for temperature
merge_health['max_temperature'] = None
for index, row in merge_health.iterrows():
    stations = Stations()
    stations = stations.region(row['ISO_2'])
    station = stations.fetch(5) 
    start = datetime(row['Start Year'],row['Start Month'],row['Start Day'])
    end = datetime(row['End Year'],row['End Month'],row['End Day'])
    data = Daily(station, start, end)
    data = data.fetch()
    tmax = data.groupby('time', as_index=False)['tmax'].max().max()
    merge_health.at[index, 'max_temperature'] = tmax[0]
merge_health.head()
```

### Drop the rows with missing max temperature

```python
df_death=merge_health.dropna()
df_death.isna().sum()
```

```python
#drop the nusiance col
df_death.drop(columns=['Start Year','End Year','Start Month','End Month','Start Day','End Day'],inplace=True)
df_death.to_csv('/Users/apple/Desktop/MDA-1/Data/df_death.csv',index=False)
```

# 2.Predict the occurrence of a heatwave.

```python
#Collecting temperature data from the Netherlands for the past decade.
stations = Stations()
stations = stations.region("BE")
station = stations.fetch(1) 
start = datetime(2013,1,1)
end = datetime(2023,6,30)
t_be = Daily(station, start, end)
t_be= t_be.fetch()
t_be.reset_index(inplace=True)
tmax_be=t_be[['time','tmax']]
tmax_be
```

```python
date = pd.DataFrame({'time': pd.date_range(start='2013-01-01', end='2023-06-30')})
df_tem = date.merge(tmax_be, on='time', how='left')
df_tem
```

## Dealing with missing value

```python
df_tem.isna().sum()
```

```python
df_tem["tmax"] = df_tem["tmax"].interpolate(method='linear')
df_tem.head()
```
